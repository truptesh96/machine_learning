# Time series Furniture Sale

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("furniture-data.csv", encoding = "ISO-8859-1", index_col = 'Row ID')
df.head()
```

```python
print("Shape:",df.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")
import itertools
import statsmodels.api as sm
```

```python
import matplotlib
matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['text.color'] = 'k'
```

```python
# Dropping unwanted variables
cols = ['Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Quantity', 'Discount', 'Profit']
df.drop(cols, axis=1, inplace=True)
df = df.sort_values('Order Date')
df.isnull().sum()
```

```python
# Creating new Data Framw with 'Order Date' and 'Sales'
df = df.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
print("Initial Date:",df['Order Date'].min())
print("End Date:",df['Order Date'].max())
```

```python
# Before changing datatype of Order Date as date
df.head()
```

```python
# Reset index after conve 'Order Date' to Datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)
```

```python
# After changing datatype of Order Date as date and reset index
df.head()
```

```python
y = df['Sales'].resample('MS').mean()
# For month start(MS) frequency
# For month (M) frequency

y.head()
```

```python
# Originial Data Representation
y.plot(figsize=(15, 6))
plt.title("Originial Data after Imputation via reampling")
plt.xlabel("Time")
plt.ylabel("Order Data")
plt.style.use('seaborn-whitegrid')
plt.show()
```

## Checking Stationarity

```python
# We are using adfuller test for Checking Stationarity
from pandas import Series
from statsmodels.tsa.stattools import adfuller
result = adfuller(y)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

# adfuller Test for stationarity

<b>Null Hypothesis: </b> It has some time dependent structure(series is non-stationary).
<br/>
<b>Alternative Hypothesis:</b> It has no time dependent structure(series is stationary).
<br/>

Here, we can see that -5.191070 < -2.944.<br/>
So, at 95% confidence level we can say that series is not stationary.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(y)

plt.plot(y, label = 'Original')
plt.legend(loc = 'best')

trend = decomposition.trend
plt.show()
plt.plot(trend, label = 'Trend')
plt.legend(loc = 'best')

seasonal = decomposition.seasonal
plt.show()
plt.plot(seasonal, label = 'Seasonal')
plt.legend(loc = 'best')

residual = decomposition.resid
plt.show()
plt.tight_layout()
plt.plot(residual, label = 'Residual')
plt.legend(loc='best')
```

## Select Parameter

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
pdq
```

```python
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
seasonal_pdq
```

```python
print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
from pylab import rcParams
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y, order=param, seasonal_order=param_seasonal, enforce_stationarity=False,  
                                            enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

NOTE: Select a model for which AIC value is small. <b>ARIMA(1, 1, 1)x(1, 1, 0, 12)12 - AIC:297.7875439545957</b> is the lowest AIC so we will take it for p d and q value.

```python
# Run SARIMAX model with best order and seasonal_order
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 0, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

# Validating Forecasts

```python
#set forecasts to start at 2017–01–01 to the end of the data to forecast
pred = results.get_prediction(start=pd.to_datetime('2017-01-01'), dynamic=False)
pred_ci = pred.conf_int()
```

```python
pred_ci
```

```python
## Plotting prediction and actual value for sales. Test model on testing Data
ax = y['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1])
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.tight_layout()
plt.show()
```

# Calculating RMSE  (Root Means Squared Error)

```python
y_forecasted = pred.predicted_mean
y_actual = y['2017-01-01':]
mse = ((y_forecasted - y_actual) ** 2).mean()
print('The Root Mean Squared Error of our forecasts is {}'.format(round(np.sqrt(mse), 2)))
```

```python
# Prediction for one year
pred_uc = results.get_forecast(steps=12)
pred_ci = pred_uc.conf_int()
ax = y.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```
